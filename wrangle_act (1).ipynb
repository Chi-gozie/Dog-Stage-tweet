## Issue #12: dog stage

## Define: Merge Doggo/Floofer/Pupper/ Puppo data into a Single Column.

## Codes

```python
archive_new.doggo.replace('None', '', inplace=True) 
archive_new.floofer.replace('None', '', inplace=True) 
archive_new.pupper.replace('None', '', inplace=True) 
archive_new.puppo.replace('None', '', inplace=True)
```

```python
archive_new['dog_stage'] = archive_new.doggo + archive_new.floofer + archive_new.pupper + archive_new.puppo
```

```python
archive_new.loc[archive_new.dog_stage == 'doggopupper', 'dog_stage'] = 'doggo, pupper' 
archive_new.loc[archive_new.dog_stage == 'doggopuppo', 'dog_stage'] = 'doggo, puppo' 
archive_new.loc[archive_new.dog_stage == 'doggofloofer', 'dog_stage'] = 'doggo, floofer'
```

```python
archive_new.loc[archive_new.dog_stage == '', 'dog_stage'] = np.nan
```
